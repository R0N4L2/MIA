#1. What is the average BACDistance (in Km)? Rename the ouptut column to "avg_bac_distance."

```python
sql="""
Select EstablishmentName
, avg([BACDistance (in Km)]) as avg_bac_distance
from school_one
"""
df_sql=pd.read_sql_query(sql,engine)
df_sql.head()
```

# 2. ¬øCu√°l es el BACDistance promedio (en km) seg√∫n nivel educativo del colegio?  Ordena los resultados de promedio mayor a promedio menor.

```python
sql="""
Select SchoolLevel
, avg([BACDistance (in Km)]) 
from school_one
order by SchoolLevel DESC
"""
df_sql=pd.read_sql_query(sql,engine)
df_sql.head()
```

# 3. Repite la consulta anterior, pero solo muestra los resultados donde la distancia promedio sea menor a 10 km.


```python
sql="""
Select SchoolLevel
, avg([BACDistance (in Km)])<10
from school_one
order by SchoolLevel DESC
"""
df_sql=pd.read_sql_query(sql,engine)
df_sql.head()
```
